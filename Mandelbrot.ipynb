# Define Mandelbrot function and function to check convergence

```python
def f(z,c):
    return z**2+c

def check_conv(z0,c,thresh,N):
    temp, i = f(z0,c), 0
    mask = np.zeros(len(c))
    while i < N:
        temp = f(temp,c)
        mask += (mask==0).astype(int)* i*(np.abs(temp)>thresh).astype(int)
        temp[mask>0] = 0
        i+=1
    
    return mask
```

# Define range and matrix of $c$ values

```python
size = 1
xmin, ymin = -2., -0.5
npts = 2500
xdiscr, ydiscr = np.linspace(xmin,xmin + size,npts), np.linspace(ymin,ymin + size,npts)
x, y = np.meshgrid(xdiscr,ydiscr)
cmat = np.array([x + 1J * y]).flatten() #np.array([complex(i,j) for i in discr for j in discr])
```
